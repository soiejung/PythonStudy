# 다중조건

- and, or, not
 - 판다스에서는 &(앰퍼샌드), | (파이프라인), ~(틸드) 문자를 사용해서 표현합니다.
 - R의 문법을 그대로 따와서 판다스가 물려받음

````python
# 장미반에서 75점 이상 맞은 사람만 찾아주세요
df.columns
````

````python
df.loc[(df.반 == '장미') & (df.테스트점수 >= 75)]
# ValueError: The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().
````

````python
(df.반 == '개나리')
````

````python
df.loc[ (df.테스트점수 >= 75) & (df.반 == '장미') | (df.반 == '개나리')]
````

````python
# 점수가 75점 이상 맞은 사람 중에 장미나 개나리반 사람을 검색해주세요
df.loc[ (df.테스트점수 >= 75) & ((df.반 == '장미') | (df.반 == '개나리'))]
````

````python
# 점수가 75점 이상 맞은 사람 중에 장미나 개나리반이 아닌 사람을 검색해주세요
df.loc[ (df.테스트점수 >= 75) & ~((df.반 == '장미') | (df.반 == '개나리'))]
````

````python
df.loc[ (df.테스트점수 >= 75) & (~(df.반 == '장미') | (df.반 == '개나리'))]
````

````python
df.loc[ (df.테스트점수 >= 75) & ~(df.반 == '장미') | (df.반 == '개나리') ]
````
